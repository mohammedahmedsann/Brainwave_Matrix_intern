#### Top Manufacturers by  Revenue

```python
# grouping manufacturers by revenue generated
sales_by_company = df.groupby('Company')['Price ($)'].agg(
    Total_Revenue='sum',
    Cars_Sold='count'
).sort_values(by='Total_Revenue', ascending=False)
# Top 10 Manufacturers by Revenue
print("Top Manufacturers by Revenue:")
print(sales_by_company.head(10))

# top manufacturers by volume (cars sold)
top_by_volume = sales_by_company.sort_values(by='Cars_Sold', ascending=False)
print("\nTop Manufacturers by Sales Volume:")
print(pd.DataFrame(top_by_volume).head(10))
```

```python
# showing top manufacturers by revenue
sales_by_company.head(10)['Total_Revenue'].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Manufacturers by Revenue')
plt.ylabel('Revenue ($)')
plt.xlabel('Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
top manufacturers by sales volume
top_by_volume.head(10)['Cars_Sold'].plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('Top 10 Manufacturers by Sales Volume')
plt.ylabel('Number of Cars Sold')
plt.xlabel('Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_region = df.groupby("Dealer_Region")["Price ($)"].sum()
pd.DataFrame(sales_by_region).sort_values(by  ="Price ($)", ascending = False)
```

```python

# Group by region and total revenue
sales_by_region = df.groupby("Dealer_Region")["Price ($)"].sum()

# Convert to DataFrame and sort
sales_by_region_df = pd.DataFrame(sales_by_region).reset_index()
sales_by_region_df.columns = ['Dealer_Region', 'Total_Revenue']
sales_by_region_df = sales_by_region_df.sort_values(by="Total_Revenue", ascending=False)

# Plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(sales_by_region_df['Dealer_Region'], sales_by_region_df['Total_Revenue'], color='skyblue')
plt.title('Total Revenue by Dealer Region', fontsize=16)
plt.xlabel('Dealer Region', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    data=sales_by_region_df, 
    x='Dealer_Region', 
    y='Total_Revenue', 
    palette='viridis'
)
plt.title('Total Revenue by Dealer Region', fontsize=16)
plt.xlabel('Dealer Region', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extracting Year and Month from Date
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
yearly_sales = df.groupby('Year')['Price ($)'].sum().reset_index()
```

```python
monthly_sales = df.groupby('Month')['Price ($)'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Plotting yearly sales trends
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales['Year'], yearly_sales['Price ($)'], marker='o', color='blue')
plt.title('Yearly Sales Trends', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting monthly sales trends
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Price ($)'], marker='o', color='green')
plt.title('Monthly Sales Trends (Across All Years)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(range(1, 13))  # Months from 1 to 12
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_yearly_sales = df.groupby([df['Date'].dt.to_period('M')])['Price ($)'].sum().reset_index()
monthly_yearly_sales['Date'] = monthly_yearly_sales['Date'].dt.to_timestamp()


# Plotting seasonal sales trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_yearly_sales['Date'], monthly_yearly_sales['Price ($)'], marker='o', linestyle='-', color='purple')
plt.title('Seasonal Sales Trends (Monthly)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by Gender and total sales and count
gender_sales = df.groupby('Gender')['Price ($)'].agg(
    Total_Revenue='sum',
    Sales_Count='count'
).reset_index()
```

```python
# percentage contribution column
total_sales = gender_sales['Total_Revenue'].sum()
gender_sales['Revenue_Percentage'] = (gender_sales['Total_Revenue'] / total_sales) * 100
```

```python

# Plotting gender-wise total revenue
plt.figure(figsize=(8, 5))
plt.bar(gender_sales['Gender'], gender_sales['Total_Revenue'], color=['blue', 'pink'])
plt.title('Gender-wise Total Revenue', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.tight_layout()
plt.show()

# Plotting gender-wise sales count
plt.figure(figsize=(8, 5))
plt.bar(gender_sales['Gender'], gender_sales['Sales_Count'], color=['blue', 'pink'])
plt.title('Gender-wise Sales Count', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Plotingt gender-wise revenue percentage
plt.figure(figsize=(8, 8))
plt.pie(
    gender_sales['Revenue_Percentage'], 
    labels=gender_sales['Gender'], 
    autopct='%1.1f%%', 
    colors=['blue', 'pink'], 
    startangle=140
)
plt.title('Gender-wise Revenue Contribution', fontsize=16)
plt.show()
```
